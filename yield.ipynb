```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv', usecols=['Observation Date', 'Series Display Name', 'Observation Value'])
```

```python
df['Observation Date'] = pd.to_datetime(df['Observation Date'])
```

```python
df = df.loc[df['Observation Date'] >= '2023-08-01']
```

```python
df = df.loc[df['Series Display Name'] == ". Treasury Bills' Cut-off Yield 3-Months"]
```

```python
df.dtypes
```

```python
df['Observation Date'].min()
```

```python
df['Observation Date'].max()
```

```python
df['Observation Value'].mean()
```

```python
# Plot the obvseration value against observation date
df.plot(x='Observation Date', y='Observation Value', title="3-Months Treasury Bills' Cut-off Yield", ylabel='Cut-off Yield (%)', xlabel='Observation Date')
```
